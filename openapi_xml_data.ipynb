```python
import requests
```

```python
req_url = 'http://openapi.work.go.kr/opi/opi/opia/wantedApi.do'
```

```python
parameters = {
    'authKey':'WNKMBKIW3GCNURZ1ABGNL2VR1HJ',
    'callTp':'L',
    'returnType':'XML',
    'startPage':1,
    'display':100,
    'region':'26000|27000'
}
```

```python
req = requests.get(req_url, params=parameters)
```

```python
content = req.text
content
```

```python
from xml.etree import ElementTree as ET
```

```python
root = ET.fromstring(content)
```

```python
wanted = root.findall('.//wanted')
```

```python
s_data = []
for want in wanted:
    company = want.find('company').text
    title = want.find('title').text
    career = want.find('career').text + ' / ' 
    career += want.find('minEdubg').text
    
    basicAddr = want.find('basicAddr').text
    closeDt = want.find('closeDt').text
    s_data.append([company, title, career, basicAddr, closeDt])
```

```python
from flask import Flask, render_template, request
import tensorflow.compat.v1 as tf
import numpy as np
import datetime
```

```python
@app.route('/') 
def index():
    now = datetime.now()
    return render_template('hoon.html',s_data=now) 


if __name__ == '__main__': 
    app.run(debug=True)
```
